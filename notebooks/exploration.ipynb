# Sales Forecasting Exploration

This notebook loads the bundled sample dataset, performs preprocessing, and generates a quick exploratory plot.


```python
import sys
from pathlib import Path
import pandas as pd

project_root = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from src.preprocessing import preprocess_sales_data

data_path = project_root / 'data' / 'sample_sales.csv'
raw_df = pd.read_csv(data_path)
clean_df, frequency = preprocess_sales_data(raw_df)
clean_df.head()
```
